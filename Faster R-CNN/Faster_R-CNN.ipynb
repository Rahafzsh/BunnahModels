```python
import os
import shutil

def separate_data(data_dir):
  """
  Separates images and labels (XML) from a combined directory.

  Args:
    data_dir: Path to the directory containing combined data.
  """
  # Create directories for separated images and labels
  image_dir = os.path.join(data_dir, 'images')
  label_dir = os.path.join(data_dir, 'labels')
  os.makedirs(image_dir, exist_ok=True)
  os.makedirs(label_dir, exist_ok=True)

  # Loop through files in the data directory
  for filename in os.listdir(data_dir):
    if filename.endswith('.jpg'):
      image_path = os.path.join(data_dir, filename)
      shutil.move(image_path, image_dir)
    elif filename.endswith('.xml'):
      label_path = os.path.join(data_dir, filename)
      shutil.move(label_path, label_dir)
    else:
      print(f"Skipping unknown file: {filename}")  # Handle unknown files (optional)

# Example usage (replace path as needed)
data_dir = '/content/custom_data/train'
separate_data(data_dir)
```

```python
import matplotlib.pyplot as plt
import glob as glob
```

```python
results_dir_path = '/content/fastercnn-pytorch-training-pipeline/outputs/training/custom_training'
valid_images = glob.glob(f"{results_dir_path}/*.jpg")

for i in range(3):
    plt.figure(figsize=(10, 7))
    image = plt.imread(valid_images[i])
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python


def display_validation_images(results_dir_path):
  """Displays up to three validation images found in the specified directory.

  Args:
      results_dir_path (str): Path to the directory containing validation images.

  Returns:
      None
  """

  valid_images = glob.glob(f"{results_dir_path}/*.jpg")
  num_images_to_display = min(3, len(valid_images))  # Limit to max 3 images

  for i in range(num_images_to_display):
    try:
      image = plt.imread(valid_images[i])
      plt.figure(figsize=(10, 7))
      plt.imshow(image)
      plt.axis('off')
      plt.title(f"Validation Image {i+1}")  # Add informative title
      plt.show()
    except IndexError:
      print(f"Skipping image {i+1} as it doesn't exist in the list.")

# Example usage
results_dir_path = '/content/fastercnn-pytorch-training-pipeline/outputs/training/custom_training'
display_validation_images(results_dir_path)
```
